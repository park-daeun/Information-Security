```python
tbl = []
for i in range(0,256):
    tbl.append(str(bin(i)[2:]).zfill(8))

def byteToString(inp):
    i = inp&0xff
    return tbl[i]

def byteArrayToString(inp):
    res = ""
    i = 0
    while i < len(inp):
        res += byteToString(inp[i])
        i += 1
    return res

#텍스트를 num개수마다 나누고 그걸 binary로 바꿔 list에 저장
def split_text_to_bin(text,num):
    i = 0
    res_block = []
    while i < len(text):
        data = []
        for j in text[i:i+num]:
            data.append(ord(j))
        if len(plain_text) % 8 != 0:
            if i > len(text) - num:
                while len(data) < 8:
                    data.append(0)
        res_block.append(byteArrayToString(data))
        i += num
    return res_block

#DES Key Scheduling
def key_scheduling(K): 
    #텍스트를 bit로
    key_ascii = []
    for i in K:
        key_ascii.append(ord(i))
    key_bit = byteArrayToString(key_ascii)

    #PC1
    res = ""
    for i in range(0,56):
        res += key_bit[PC1[i]]
        
    C = res[:28]
    D = res[28:]

    ##Making Sub Keys
    subkey = []
    for i in range(0,16):
        C = C[Shift[i]:] + C[:Shift[i]]
        D = D[Shift[i]:] + D[:Shift[i]]
        res_key = C+D
        res = ""
        for j in range(0,48):
            res += res_key[PC2[j]-1]
        subkey.append(res)
        
    return subkey

def Initial_permutation(text):
    after_IP = ""
    for i in range(0,64):
        after_IP += text[IP[i]-1]
    L = after_IP[:32]
    R = after_IP[32:]
    return L,R

#F-box
def F(R,SK):
    #Expansion R
    expanded_R = ""
    for i in range(0,len(E)):
        expanded_R += R[E[i]-1]       
    temp = str(bin(int(SK,2) ^ int(expanded_R,2))[2:]).zfill(48)
    
    #S-box
    s_box_input = []
    i = 0
    while i < len(temp):
        s_box_input.append(temp[i:i+6])
        i += 6
    
    s_box_output = []
    for j in range(0,len(s_box_input)):
        row = int((s_box_input[j][0]+s_box_input[j][5]),2)
        col = int((s_box_input[j][1]+s_box_input[j][2]+s_box_input[j][3]+s_box_input[j][4]),2)
        
        s_box_output.append(str(bin(SBOX[j][row][col])[2:]).zfill(4))
    
    output_str = ""
    for k in s_box_output:
        output_str += k
    
    return output_str

def feistel_round(L,R,SK):
    temp = R
    after_F = F(R,SK)
    
    #Permutaion(Another Bit Shuffling)
    permute = ""
    for j in range(0, len(P)):
        permute += after_F[P[j]-1]
            
    R = str(bin(int(L,2) ^ int(permute,2))[2:]).zfill(32)
    L = temp
    
    return L,R

def final_permutation(bits):
    res = ""
    for j in range(0,64):
        res += bits[FP[j]-1]
    return res

#block의 값들을 텍스트로 합치기
def block_to_text(block):
    text = ""
    for i in range(0,len(block)):
        res = ""
        for j in range(0,64,8):
            bin_text = block[i][j:j+8]
            res+= chr(int(bin_text,2))
        text += res
    return text
```
